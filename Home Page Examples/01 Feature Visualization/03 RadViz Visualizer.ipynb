# RadViz Visualizer
```RadViz``` は多変量データ視覚化アルゴリズムであり、円の円周上に各フィーチャの次元を均一にプロットし、次に中心から各円弧までの軸上の値を正規化するように円の内側に点をプロットします。 。
このメカニズムにより、円に簡単に収まる限り多くの次元が可能になり、視覚化の次元が大幅に拡張されます。

データ サイエンティストは、この方法を使用してクラス間の分離可能性を検出します。
例えば。 機能セットから学ぶ機会はあるのでしょうか、それともノイズが多すぎるのでしょうか?

データに欠損値 (```numpy.nan```) を含む行が含まれている場合、それらの欠損値はプロットされません。 言い換えれば、データの全体像を把握できない可能性があります。 ```RadViz``` は ```DataWarning``` を発生させて、不足しているパーセントを通知します。

この警告を受け取った場合は、代入戦略を検討することをお勧めします。 [scikit-learn Imputer](http://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html) から始めるのが良いでしょう。

| Kind | Data type |
| ---- | ---- |
| Visualizers | ```RadialVisualizer``` |
| Quick Methods | ```radviz()``` |
| Models | Classification, Regression |
| Workflow | Feature Engineering |

[API Reference](https://www.scikit-yb.org/en/latest/api/features/radviz.html#module-yellowbrick.features.radviz)

`````python
from yellowbrick.datasets import load_occupancy
from yellowbrick.features import RadViz

# Load the classification dataset
X, y = load_occupancy()

# Specify the target classes
classes = ["unoccupied", "occupied"]

# Instantiate the visualizer
visualizer = RadViz(classes=classes)

visualizer.fit(X, y)           # Fit the data to the visualizer
visualizer.transform(X)        # Transform the data
visualizer.show()              # Finalize and render the figure
`````

回帰の場合、```RadViz```` ビジュアライザは、個別の色ではなく、色のシーケンスを使用してターゲット情報を表示する必要があります。

## Quick Method
上記と同じ機能は、関連するクイック メソッド ```radviz``` を使用して実現できます。
このメソッドは、関連付けられた引数を使用して ```RadViz``` オブジェクトを構築し、それを適合させてから、(オプションで) すぐに視覚エフェクトを表示します。

`````python
from yellowbrick.features.radviz import radviz
from yellowbrick.datasets import load_occupancy

#Load the classification dataset
X, y = load_occupancy()

# Specify the target classes
classes = ["unoccupied", "occupied"]

# Instantiate the visualizer
radviz(X, y, classes=classes)
`````
